```python
import pandas as pd
```

```python
df = pd.read_csv('data/frog_tongue_adhesion.csv', comment='#')
```

```python
df.head()
```

```python
#a)
df.loc[df['adhesive strength (Pa)']< -2000,'impact time (ms)']
```

```python
#b) Extract the impact force and adhesive force for all of Frog II's strikes.
df.loc[df['ID']== 'II',('impact force (mN)','adhesive force (mN)')]
```

```python
# cï¼‰
df.loc[df['ID']== ('III' or 'IV'),('time frog pulls on target (ms)','adhesive force (mN)')]
```

```python
# 2
#a) standard deviation of the impact forces for each frog.
grouped_by_ID=df.groupby('ID')
```

```python
grouped_by_ID['impact force (mN)'].std()
```

```python
# b) 
import numpy as np
def coeff_of_var(data):
    return np.std(data) / np.mean(data)
```

```python
grouped_by_ID['impact force (mN)','adhesive force (mN)'].agg(coeff_of_var)
```

```python
df_mini=grouped_by_ID['impact force (mN)','adhesive force (mN)'].agg([coeff_of_var,np.std,np.mean,np.median])
df_mini
```
