```python
df_frog = pd.DataFrame(index=['I', 'II', 'III', 'IV'],
                       data={'age': ['adult', 'adult', 'juvenile', 'juvenile'],
                             'SVL (mm)': [63, 70, 28, 31],
                             'weight (g)': [63.1, 72.7, 12.7, 12.7],
                             'species': ['cross', 'cross', 'cranwelli', 'cranwelli']})
df_frog
```

```python
df = pd.read_csv('data/frog_tongue_adhesion.csv', comment='#')
df.head()
```

```python
for i in ['I','II','III','IV']:
    for j in ['age','SVL (mm)','weight (g)','species']:
        df.loc[df['ID']==i,j]=df_frog.loc[i,j]
```

```python
df
```

```python
# bad idea
def label_SVL (row):
    if row['ID'] == 'I' :
        return 63
    if row['ID'] == 'II' :
        return 70
    if row['ID'] == 'III' :
        return 28
    if row['ID'] == 'IV' :
        return 31
    return 'Other'
```

```python
def label_species (row):
    if row['ID'] == 'I':
        return 'cross'
    if row['ID'] == 'II':
        return 'cross'
    if row['ID'] == 'III' :
        return 'cranwelli'
    if row['ID'] == 'IV':
        return 'cranwelli'
    return 'Other'
```

```python
def label_age (row):
    if row['ID'] == 'I':
        return 'adult'
    if row['ID'] == 'II':
        return 'adult'
    if row['ID'] == 'III' :
        return 'juvenile'
    if row['ID'] == 'IV':
        return 'juvenile'
    return 'Other'
```

```python
df['age']=df.apply (label_age ,axis=1)
df['species']=df.apply (label_species ,axis=1)
df['SVL']=df.apply (label_SVL ,axis=1)
```

```python
alt.Chart(df
    ).mark_point(
    ).encode(
        x='adhesive force (mN):Q',
        y='impact force (mN):Q',
        color='age:N'
)
```

```python
# 3.3
df_iris=pd.read_csv('data/fisher_iris.csv')
```

```python
df_iris.head()
```

```python
brush = alt.selection(type='interval')

tick_axis = alt.Axis(labels=False, domain=False, ticks=False)
tick_axis_notitle = alt.Axis(labels=False, domain=False, ticks=False, title='')

points = alt.Chart(df_iris).mark_point().encode(
    x=alt.X('petal width (cm):Q', axis=alt.Axis(title='')),
    y=alt.Y('petal length (cm):Q', axis=alt.Axis(title='')),
    color=alt.condition(brush, 'species:N', alt.value('grey'))
).add_selection(
    brush
)

x_ticks = alt.Chart(df_iris).mark_tick().encode(
    alt.X('petal width (cm):Q', axis=tick_axis),
    alt.Y('species:N', axis=tick_axis_notitle),
    color=alt.condition(brush, 'species:N', alt.value('lightgrey'))
).add_selection(
    brush
)

y_ticks = alt.Chart(df_iris).mark_tick().encode(
    alt.X('species:N', axis=tick_axis_notitle),
    alt.Y('petal length (cm):Q', axis=tick_axis),
    color=alt.condition(brush, 'species:N', alt.value('lightgrey'))
).add_selection(
    brush
)

y_ticks | (points & x_ticks)
```

```python
alt.Chart(df_iris).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='species:N'
).properties(
    width=200,
    height=200
).repeat(
    row=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'],
    column=['petal width (cm)','petal length (cm)','sepal width (cm)','sepal length (cm)']
).interactive()
```
